```python
data_impact = pd.read_csv('impact factor.csv')
data_impact
```

```python
d = ['Name', 'Temp multiplier']

df = pd.DataFrame(columns=d)
```

```python
for label in list_label:
    line_impact_factor = data_impact[data_impact['label'] == retrieve_name(label)[0]]
    impact_factor = line_impact_factor.iloc[0]['delta']
    for contry in label:
        new_df = pd.DataFrame(data={'Name' : [contry], 'Temp multiplier': [impact_factor]})
        df = pd.concat([df, new_df])
```
