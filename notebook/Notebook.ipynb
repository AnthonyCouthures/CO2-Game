```python
import numpy as np
#function [u, y, z]  = Utility(a, s, alpha, beta, delta, y0, z0)
def Utility(a, s, alpha, beta, gamma, delta, kappa, y0, z0, N, T, eta, z1ref, Ay, Az, By,  Bz, emax, amax, fex0, fex1, tf) :
    ###################################################################################################
    anmax = amax*np.ones([N, 1])
    Bn = emax / anmax
    ############
    c1 = By[0][0]

    ############
    z = (z0)                         # CO2 concentration vector
    print("z:",z)
    print("z0:",z0)

    t = 0
    #M1 = 0
    MM1 = np.zeros(T)
    print(MM1[1])
    while t<T-1 : 
       #M1 = Bz'*Az*z0
        print(Bz.T @ Az @ z0)
        MM1[t + 1] = Bz.T @ Az @ z0
        print("o",s[t] + a[t])
        z0 = Az@z0 + Bz@(s[t] + a[t])
        z = [z, z0]
        t += 1  
    
    ############                       
    forcoth = np.zeros([1, T])             # Exogeneous forcing for other greenhouse gases
    F = np.zeros([1, T])
    z1 = [1, 0, 0]*z                   # CO2 atmospheric concentration  
    for t in range(T-1):
        forcoth[t]  =  fex0 + min(fex1-fex0, ((fex1-fex0)/tf)*(t-1))
        F[t] = eta*(np.log(z1[t]/z1ref)/np.log(2)) + forcoth[t]
        #F[t] = forcoth[t]-eta*log(z1ref)/log(2) + eta*log(z1[t])/log(2)

    ############
    y = (y0)                         # Temperature vector
    t = 0
    #M = 0
    MM = np.zeros([1, T])

    while t<T-1:
       #M = M + c1*forcoth[t]-eta*c1*log(z1ref) + 1/c1*By'*Ay*y0
       MM[t + 1] = MM[t] + c1*forcoth[t]-eta*c1*np.log(z1ref) + 1/c1*By.T * Ay * y0
       y0 = Ay*y0 + By*(F[t])
       y = [y, y0]
       t += 1   
    
    ############
    A = 2 * emax^2 / anmax[N]*a - emax**2/(anmax[N]**2)*a**2        # Benefits function
    B = np.array([1, 0]) * y                                             # Damage function
    C = np.array([1, 0] * y)**2                                        # Damage function
    u = alpha*A - beta*B - delta*C                              # Utility function 
    D = beta*B + delta*C                                      # Damage function
    u1 = alpha*(Bn[N])**2 * a**2 + beta*(Bn[N]) * a + gamma - delta * B - kappa * C

    return u1, y, z
```
