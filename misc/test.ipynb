```python
data_impact = pd.read_csv('data\impact factor.csv')
data_impact
```

```python
d = ['Entity', 'Impact Factor']

df = pd.DataFrame(columns=d)
```

```python
df
```

```python
for label in list_label:
    line_impact_factor = data_impact[data_impact['label'] == retrieve_name(label)[0]]
    impact_factor = line_impact_factor.iloc[0]['delta']
    for contry in label:
        new_df = pd.DataFrame(data={'Entity' : [contry], 'Impact Factor': [impact_factor]})
        df = pd.concat([df, new_df])
```

```python
df = df.sort_values('Entity').reset_index(drop=True)
```

```python

data = pd.read_csv('test.csv')
data.drop(['Unnamed: 0'], axis=1, inplace=True)
data_merged = pd.merge(data, df, on=['Entity'], how='left').reset_index(drop=True)
```
